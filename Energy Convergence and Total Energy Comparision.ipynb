```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

```

```python
myAxisLabelSize=32
myTickLabelSize=24
myLegendLabelSize=24
myTitleSize=30
```

```python
data = np.genfromtxt("Energy_pi15.dat")
data1 = np.genfromtxt("Energy_2pi.dat") #old QuasistaticBrownian.cpp, all others are the editted versions
data2 = np.genfromtxt("Energypi2.dat")
data3 = np.genfromtxt("Energy0pi.dat")

```

```python
totalEnergy = data[:,1]
totalEnergyDropLast = totalEnergy[:-1]
totalEnergyDropFirst = totalEnergy[1:]
totalEnergyDiff = np.abs(totalEnergyDropLast - totalEnergyDropFirst)
totalEnergyDiffFrac = totalEnergyDiff / totalEnergyDropLast

coatEnergy = data[:,3]
coatEnergyDropLast = coatEnergy[:-1]
coatEnergyDropFirst = coatEnergy[1:]
coatEnergyDiff = np.abs(coatEnergyDropLast - coatEnergyDropFirst)
coatEnergyDiffFrac = coatEnergyDiff / coatEnergyDropLast

totalEnergy_new = data1[:,1]
totalEnergyDropLast_new = totalEnergy_new[:-1]
totalEnergyDropFirst_new = totalEnergy_new[1:]
totalEnergyDiff_new = np.abs(totalEnergyDropLast_new - totalEnergyDropFirst_new)
totalEnergyDiffFrac_new = totalEnergyDiff_new / totalEnergyDropLast_new

coatEnergy_new = data1[:,3]
coatEnergyDropLast_new = coatEnergy_new[:-1]
coatEnergyDropFirst_new = coatEnergy_new[1:]
coatEnergyDiff_new = np.abs(coatEnergyDropLast_new - coatEnergyDropFirst_new)
coatEnergyDiffFrac_new = coatEnergyDiff_new / coatEnergyDropLast_new

totalEnergy_new2 = data2[:,1]
totalEnergyDropLast_new2 = totalEnergy_new2[:-1]
totalEnergyDropFirst_new2 = totalEnergy_new2[1:]
totalEnergyDiff_new2 = np.abs(totalEnergyDropLast_new2 - totalEnergyDropFirst_new2)
totalEnergyDiffFrac_new2 = totalEnergyDiff_new2 / totalEnergyDropLast_new2

coatEnergy_new2 = data2[:,3]
coatEnergyDropLast_new2 = coatEnergy_new2[:-1]
coatEnergyDropFirst_new2 = coatEnergy_new2[1:]
coatEnergyDiff_new2 = np.abs(coatEnergyDropLast_new2 - coatEnergyDropFirst_new2)
coatEnergyDiffFrac_new2 = coatEnergyDiff_new2 / coatEnergyDropLast_new2

totalEnergy_new3 = data3[:,1]
totalEnergyDropLast_new3 = totalEnergy_new3[:-1]
totalEnergyDropFirst_new3 = totalEnergy_new3[1:]
totalEnergyDiff_new3 = np.abs(totalEnergyDropLast_new3 - totalEnergyDropFirst_new3)
totalEnergyDiffFrac_new3 = totalEnergyDiff_new3 / totalEnergyDropLast_new3

coatEnergy_new3 = data2[:,3]
coatEnergyDropLast_new3 = coatEnergy_new3[:-1]
coatEnergyDropFirst_new3 = coatEnergy_new3[1:]
coatEnergyDiff_new3 = np.abs(coatEnergyDropLast_new3 - coatEnergyDropFirst_new3)
coatEnergyDiffFrac_new3 = coatEnergyDiff_new3 / coatEnergyDropLast_new3





```

```python
#Coating Energy #
fig = plt.gcf()
ax = plt.gca()
gridwidth=0.5
gridcolor=(0.8,0.8,0.8)

ax.plot(data[:,0][:-1], coatEnergyDropLast, marker='x', label='Total energy', color='r')
ax.plot(data1[:,0][:-1], coatEnergyDropLast_new, marker='x', label='Total energy', color='g')
ax.plot(data2[:,0][:-1], coatEnergyDropLast_new2, marker='x', label='Total energy', color='b')
ax.plot(data3[:,0][:-1], coatEnergyDropLast_new3, marker='x', label='Total energy', color='orange')
```

```python
#Glass Convergence #
fig = plt.gcf()
ax = plt.gca()
gridwidth=0.5
gridcolor=(0.8,0.8,0.8)

ax.plot(data[:,0][:-1], totalEnergyDiffFrac, marker='x', label='Total energy', color='r')

ax.plot(data1[:,0][:-1], totalEnergyDiffFrac_new, marker='x', label='Total energy', color='g')
ax.plot(data2[:,0][:-1], totalEnergyDiffFrac_new2, marker='x', label='Total energy', color='b')


```
